```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
df = pd.read_csv("/content/diamond_data_merged_with_other_variables.csv")
#print(df.shape)
#print(df.head)

#df.drop(columns=['date'])

print(df.head)
print(df.dtypes)

df['date'] = pd.Categorical(df['date'])
df['date'] = df.date.cat.codes

#split data set into train and test
x_train=df.sample(frac=0.8,random_state=200) #random state is a seed value
x_test=df.drop(x_train.index)


#print(x_train.shape)
#print(x_test.shape)

y_train = x_train["diamond price"]
y_test = x_test["diamond price"]

#y_train.drop("diamond price")

#print(y_train.shape)
#print(y_test.shape)
```

Keras Layers start here

```python
model = keras.Sequential([
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(32, activation='relu'),
      layers.Dense(1)
  ])

model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
```

```python
#x_train.drop[""]
x_train = tf.convert_to_tensor(
    x_train, dtype=tf.float32, dtype_hint=None, name=None
)

y_train = tf.convert_to_tensor(
    y_train, dtype=tf.float32, dtype_hint=None, name=None
)

model.fit(x_train,y_train,epochs=1000)
```
